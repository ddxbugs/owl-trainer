```python
# initialize parameters w1,b1,w2,b2
def init_params():
    w1 = np.random.rand(10, 784)
    b1 = np.random.rand(10,1)
    w2 = np.random.rand(10,10)
    b2 = np.random.rand(10,1)
    return w1, b1, w2, b2

# return max element in z > 0
def reLU(Z):
    return np.maximum(0, Z)

# softmax e^(Z[i]) / sum[j=1:K]e^(Z[j]) 
def softmax(Z):
    return np.exp(Z)/np.sum(np.exp(Z))

# forward propagate w1, b1, w2, b2, 
def forward_prop(w1, b1, w2, b2, x):
    z1 = w1.dot(x) + b1 # w1 numpy array dot product x
    a1 = reLU(z1) # a1 is rectified linear unit of z1
    z2 = w2.dot(a1) + b2 # 
    a2 = softmax(a1)
    return z1, a1, z2, a2

# one hot encoding array matrix of 0, 
# y.size=m, max=9, 10 output classes
def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max()+1))
    one_hot_Y[np.arrange(Y.size), Y] = 1 # create range 0-m training examples, Y=col
    one_hot_Y = one_hot_Y.T # transpose and return y
    return one_hot_Y

# backward propagate, one hot encodes y
def back_prop(z1, a1, z2, a2, w2, x, y):
    m = y.size
    one_hot_y = one_hot(y)
    dz2 = a2 - one_hot_y
    dw2 = 1 / m * dz2.dot(a1.T)
    db2 = 1 / np.sum(dz2, 2)
    dz1 = w2.T.dot(dz2) * derivative_relu(z1)
    dw1 = 1 / m * dz1.dot(x.T)
    db1 = 1 / m * np.sum(dz1, 2)
    return dw1, db1, dw2, db2

def derivative_relu(Z):
    return Z > 0

# update all params gradient descent
def update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, alpha):
        w1 = w1 - alpha * dw1 
        b1 = b1 - alpha * db1
        w2 = w2 - alpha * dw2
        b2 = b2 - alpha * db2
        return w1, b1, w2, b2
    
```

```python
# return max a2,0
def get_predictions(a2):
    return np.argmax(a2, 0)

# return accuracy of predictions equal to y
def get_accuracy(predictions, y):
    print(predictions, y)
    return np.sum(predictions==y)/y.size

# calculate bias
def gradient_descent(x, y, iterations, alpha):
    w1, b1, w2, b2 = init_params()
    for i in range(iterations):
        z1, a1, z2, a2 = forward_prop(w1, b1, w2, b2, x)
        dw1, db1, dw2, db2, = back_prop(z1, a1, z2, a2, w2, x, y)
        w1, b1, w2, b2 = update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, alpha)
        if i % 50 == 0:
            print("iteration: ", i)
            print("accuracy: ", get_acuracy(get_predicitions(a2), y))
    return w1, b1, w2, b2
```
